## Plot Graph

```python
# Set up the figure with two subplots (Loss and F1)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

for (weights, model), results in experiment_outputs.items():
    # Convert tensor to a simple list for the label
    weight_list = weights.tolist()
    label = f"{model} {weight_list}"
    
    # 1. Plot Training Loss
    train_loss = results['losses_train']['loss']
    epochs = range(1, len(train_loss) + 1)
    ax1.plot(epochs, train_loss, label=label, marker='o')

    # 2. Plot Validation F1 Score
    if 'f1' in results['losses_validation']:
        val_f1 = results['losses_validation']['f1']
        ax2.plot(epochs, val_f1, label=label, marker='s')

# Formatting the Loss plot
ax1.set_title('Training Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend(fontsize='small', loc='upper right')
ax1.grid(True)

# Formatting the F1 plot
ax2.set_title('Validation F1 Score')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('F1 Score')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
