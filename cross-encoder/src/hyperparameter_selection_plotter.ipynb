## Plot Graph

```python
"""
Graphs
- Validation loss
- F1 score
- Precision
- Recall
"""
fig, ((ax_loss, ax_f1), (ax_precision, ax_recall)) = plt.subplots(2, 2, figsize=(15, 15))

for (weights, model), results in experiment_outputs.items():
    label = f"{model} {str(weights)}"
    
    # Plot validation loss
    loss_validation = results["losses_validation"]["loss"]
    epochs = range(1, len(loss_validation) + 1)
    ax_loss.plot(epochs, loss_validation, label=label, marker="s")

    # Plot validation F1 score
    f1 = results["losses_validation"]["f1"]
    ax_f1.plot(epochs, f1, label=label, marker="s")

    # Plot validation precision
    precision = results["losses_validation"]["precision"]
    ax_precision.plot(epochs, precision, label=label, marker="s")

    # Plot validation recall
    recall = results["losses_validation"]["recall"]
    ax_recall.plot(epochs, recall, label=label, marker="s")


# Format plot of validation loss
ax_loss.set_title("Validation Loss")
ax_loss.set_xlabel("Epochs")
ax_loss.set_ylabel("Loss")
ax_loss.legend(loc=f"upper center", bbox_to_anchor=(
    0.5, -0.1), ncol=1, borderaxespad=0.)
ax_loss.grid(True)

# Format plot of validation F1 score
ax_f1.set_title("Validation F1 Score")
ax_f1.set_xlabel("Epochs")
ax_f1.set_ylabel("F1 Score")
ax_f1.legend(loc=f"upper center", bbox_to_anchor=(
    0.5, -0.1), ncol=1, borderaxespad=0.)
ax_f1.grid(True)

# Format plot of validation precision
ax_precision.set_title("Validation Precision")
ax_precision.set_xlabel("Epochs")
ax_precision.set_ylabel("Precision")
ax_precision.legend(loc=f"upper center", bbox_to_anchor=(
    0.5, -0.1), ncol=1, borderaxespad=0.)
ax_precision.grid(True)

# Format plot of validation recall
ax_recall.set_title("Validation Recall")
ax_recall.set_xlabel("Epochs")
ax_recall.set_ylabel("Recall")
ax_recall.legend(loc=f"upper center", bbox_to_anchor=(
    0.5, -0.1), ncol=1, borderaxespad=0.)
ax_recall.grid(True)

plt.title(f"Results of Hyperparameter Selection")
plt.tight_layout()
plt.show()
```
